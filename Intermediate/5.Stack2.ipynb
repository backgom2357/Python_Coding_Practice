### 중위표기식을 후위표기식으로

```python
a = "( 6 + 5 * ( 2 - 8 ) / 2 )"
a.split()
```

```python
operators = "()*/+-"
priority = {"(": 0,
            "+": 1, "-": 1,
            "*": 2, "/": 2}

def eq_tokenize(string):
    tokenized_list = []
    
    for c in string:
        
        if not tokenized_list:
            tokenized_list.append(c)
            continue
        
        if c in operators:
            tokenized_list.append(c)
            continue
            
        else:
            if tokenized_list[-1] not in operators:
                tokenized_list[-1] += c
            else:
                tokenized_list.append(c)
    
    return tokenized_list

eq = eq_tokenize(input())

def be_converted_to_postfix(l):
    
    stack = []
    result = []

    for token in l:

        if token in operators:

            if not stack:
                stack.append(token)
                continue

            if token == ")":

                while stack:
                    tmp = stack.pop()

                    if tmp == "(":
                        break

                    result.append(tmp)

                continue

            while stack:

                if  token =="(":
                    stack.append(token)
                    break

                if priority[stack[-1]]<priority[token]:
                    stack.append(token)
                    break
                else:
                    result.append(stack.pop())

        else:
            result.append(token)

    return result

be_converted_to_postfix(eq)
```

### 후위표기식 계산

```python
eq = input()

def calculate(string):
    
    postfix = be_converted_to_postfix(eq_tokenize(string))
    stack = []

    for token in postfix:

        if token not in operators:
            stack.append(token)
            continue

        tmp2 = stack.pop()
        tmp1 = stack.pop()

        stack.append(str(eval(tmp1+token+tmp2)))

    return stack.pop()

calculate(eq)
```

## 백트래킹

해를 찾는 도중에 '막히면(해가 아니면)' 되돌아가서 다시 해를 찾아가는 기법

최적화 문제, 결정 문제 해결가능

- 백트래킹 알고리즘의 특징
    - 백트래킹과 깊이 우선 탐색의 차이
        - 백트래킹
            - 어떤 노드에서 출발하는 경로가 해결책으로 이어질 것 같지 않으면 더 이상 그 경로를 따라가지 않음으로써 시도의 횟수를 줄임
            - 가지치기(Prunning)
            - 불필요한 경로의 조기 차단
            - N! 가지의 경우의 수를 가진 문제에 대해 백트래킹에 가하면 일반적으로 경우의 수가 줄어들지만 이역시 최악의 경우에는 여전히 Exponential Time을 요하므로 처리 불가능
            - 모든 후보를 검사하지 않음
        - 깊이 우선 탐색
            - 모든 경로를 추적
            - N! 가지의 경우의 수를 가진 문제에 대해 깊이 우선 탐색을 가하면 처리 불가능한 문제
            - 모든 후보를 검사
        - 백트래킹 기법
            - 어떤 노드의 유망성을 점검한 후에 유망(Promsing)하지 않다고 결정되면 그 노드의 부모로 되돌아가(Backtracking) 다음 자식 노드로 감
            - 어떤 노드를 방문하였을 때 그 노드를 포함한 경로가 해답이 될 수 없으면 그 노드는 유망하지 않다고 함
            - 반대로 해답의 가능성이 있으면 유망하다고 함
            - 가지치기(Prunning): 유망하지 않은 노드가 포함되는 경로는 더 이상 고려하지 않음
        - 백트래킹을 이용한 알고리즘의 절차
            - 상태 공간 Tree의 깊이 우선 검색을 실시
            - 각 노드가 유망한지 점검
            - 그 노드가 유망하지 않다면, 그 노드의 부모 노드로 가서 검색을 계속
        - Power Set
            - 어떤 집합의 공집합과 자기자신을 포함한 모든 부분집합
            - 구하고자 하는 어떤 집합의 원소 개수가 n일 경우 부분집합의 개수는 2^n이 나옴
        - 백트래킹 기법으로 Power Set 구하기
            - 일반적인 백트래킹 접근 방법 이용
            - n개의 원소가 들어있는 집합의 2^n개의 부분집합을 만들 때, True 또는 False값을 가지는 항목들로 구성된 n개의 리스트를 만드는 방법 이용
            - 리스트의 i번째 항목은 i번째의 원소가 부분집합의 값인지 아닌지를 나타내는 값

```python
## Power set

def backtrack(a, k, input_):
    global MAXCANDIDATES
    c = [0]*MAXCANDIDATES
    
    if k == input_:
        process_solution(a, k)
    else:
        k += 1
        ncandidates = construct_candidates(a, k, input_, c)
        for i in range(ncandidates):
            a[k] = c[i]
            backtrack(a, k, input_)
            
def construct_candidates(a, k, input_, c):
    c[0] = True
    c[1] = False
    return 2

def process_solution(a, k):
    print("(", end="")
    for i in range(k+1):
        if a[i]:
            print(i, end=" ")
    print(")")

MAXCANDIDATES = 100
NMAX = 100
a = [0]*NMAX
backtrack(a, 0, 3)
```

## 분할 정복

- 설계 전략
    - 분할(Divide) : 해결할 문제를 여러 개의 작은 부분으로 나눔
    - 정복(Conquer) : 나눈 작은 문제를 각각 해결
    - 통합(Combine) : (필요하다면) 해결된 해답을 모음

```python
def quickSort(a, begin, end):
    if begin < end:
        p = partition(a, begin, end)
        quickSort(a, begin, p-1)
        quickSort(a, p+1, end)
        
def partition(a, begin, end):
    pivot = (begin + end)//2
    L = begin
    R = end
    while L < R:
        while a[L]<a[pivot] and L<R:
            L += 1
        while a[R]>=a[pivot] and L<R:
            R -= 1
        if L < R:
            if L == pivot:
                pivot = R
            a[L], a[R] = a[R], a[L]
    a[pivot], a[R] = a[R], a[pivot]
    return R
```

### Forth

```python
operators = "+-*/"

def calculate(string):
    
    postfix = list(string.split(" "))
    stack = []
    
    try:
        for token in postfix:

            if token == ".":
                if len(stack) != 1:
                    return 'error'
                
                return stack.pop()

            if token not in operators:
                stack.append(token)
                continue

            tmp2 = stack.pop()
            tmp1 = stack.pop()

            stack.append(str(int(eval(tmp1+token+tmp2))))
           
    except:
        return 'error'

T = int(input())

for case in range(1,T+1):
    eq = input()
    result = calculate(eq)
    
    print("#%d"%case, result)
```

### 토너먼트 카드게임

```python
# T = int(input())

f = open("cases.txt", "r")
T = int(f.readline())

# T = int(input())


def divide(start, end, l):
    if end - start >= 2:
        p = (start + end) // 2
        start = divide(start, p, l)
        end = divide(p + 1, end, l)

    if l[start] - l[end] in [-2, 1, 0]:
        return start
    else:
        return end


for case in range(1, T + 1):
    # N = int(input())
    N = int(f.readline())
    cards = list(map(int, f.readline().split()))

    print("#%d" % case, divide(0, N - 1, cards) + 1)

f.close()
```

### 배열 최소 합

```python
# 순열

def p_construct_candidates(a, k, input_, c):
    in_perm = [False]*NMAX
    
    for i in range(1,k):
        in_perm[a[i]] = True
        
    ncandidates = 0
    for i in range(1,input_+1):
        if in_perm[i] == False:
            c[ncandidates] = i
            ncandidates += 1
    return ncandidates

def p_backtrack(a, k, input_):
    global MAXCANDIDATES, permutation
    c = [0]*MAXCANDIDATES
    
    if k == input_:
        permutation.append(a[1:k+1])
    else:
        k += 1
        ncandidates = p_construct_candidates(a, k, input_, c)
        for i in range(ncandidates):
            a[k] = c[i]
            p_backtrack(a, k, input_)
            
MAXCANDIDATES = 100
NMAX = 100
permutation = []
T = int(input())

for case in range(1,T+1):
    
    N = int(input())
    square = []
    
    a = [0]*NMAX
    permutation = []
    p_backtrack(a,0,N)
    
    for i in range(N):
        square.append(list(map(int,input().split())))

    sum_candidates = []
    
    for num_list in permutation:
        total = 0
        for j in range(N):
            total += square[j][num_list[j]-1]
        sum_candidates.append(total)

    print("#%d"%case, min(sum_candidates))
```
